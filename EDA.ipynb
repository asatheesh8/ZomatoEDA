### Country Distribution (Visualised)

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
country_names
```

```python
country_val=final_df.Country.value_counts().values
country_val
```

```python
## Pie Chart- Top 3 countries that uses zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

The Top 3 Countries that use ZOMATO are India (94.49%), US (4.64%) and UK (0.87%)

### Relationships

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

#### Observation

* 0 --> Not rated
* 1.0 - 2.4 --> Poor (red)
* 2.5 - 3.4 --> Average (orange)
* 3.5 - 3.9 --> Good (yellow)
* 4.0 - 4.4 --> Very Good (green)
* 4.5 - 5.0 --> Excellent (dark green)

```python
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings, hue='Rating color',palette=['lightblue','red','orange','yellow','lightgreen','green'])
```

#### Observation

* Unrated count is really high
* Maximum distribution is between 2.8 to 3.9

```python
sns.countplot(x='Rating color',data=ratings,hue='Rating color',palette=['lightblue','red','orange','yellow','lightgreen','green'])
```
